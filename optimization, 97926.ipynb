# optimization 

after analysis of all transport properties we optimize our packed bed reactor for industrial use . as, optimization that we have done is on the base of pressure drop As in case of soybean catalyst for biodiesel production we require lowest pressure drop that's we find in catalyst 1

##  importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## pressure drop values along different velocity iterations

```python
# Our simulation to PNM model b/w Pressure drop and Velocity
import numpy as np

u = np.arange(0.02,0.6,0.02)  
viscosity = 1.79*10**-5 # kg/m.s
epsilon1 = 0.37586413
epsilon2 = 0.30673585
epsilon3 = 0.24561348
epsilon4 = 0.19751556
epsilon5 = 0.14655910
dp = 0.00724 # m
# œÅ=1.257 
density = 1.025 # kg/m3

# catalyst 1
delta_p1 = 150 * u * viscosity * (1 - epsilon1)**2 / ((dp**2)* (epsilon1**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon1)) / (dp * epsilon1 **3)
dp1 = (delta_p1 + delta_p2) * 130/1000
print(dp1)

# catalyst 2
delta_p1 = 150 * u * viscosity * (1 - epsilon2)**2 / ((dp**2)* (epsilon2**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon2)) / (dp * epsilon2 **3)
dp2 = (delta_p1 + delta_p2) * 130/1000
print(dp2)

# catalyst 3
delta_p1 = 150 * u * viscosity * (1 - epsilon3)**2 / ((dp**2)* (epsilon3**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon3)) / (dp * epsilon3 **3)
dp3 = (delta_p1 + delta_p2) * 130/1000
print(dp3)

# catalyst 4
delta_p1 = 150 * u * viscosity * (1 - epsilon4)**2 / ((dp**2)* (epsilon4**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon4)) / (dp * epsilon4 **3)
dp4 = (delta_p1 + delta_p2) * 130/1000
print(dp4)

# catalyst 5
delta_p1 = 150 * u * viscosity * (1 - epsilon5)**2 / ((dp**2)* (epsilon5**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon5)) / (dp * epsilon5 **3)
dp5 = (delta_p1 + delta_p2) * 130/1000
print(dp5)

```

##  pressure drop vs velocity

```python
fig, ax = plt.subplots(figsize=(8, 5))
plt.scatter(u,dp1 , color='red', marker='*', label='Catalyst 1')
plt.scatter(u,dp2 , color='blue', marker='8', label='Catalyst 2')
plt.scatter(u,dp3 , color='orange', marker='s', label='Catalyst 3')
plt.scatter(u,dp4 , color='green', marker='p', label='Catalyst 4')
plt.scatter(u,dp5 , color='magenta', marker='x', label='Catalyst 5')
plt.xlabel('Velocity (m/s)',fontweight='bold')
plt.ylabel('Pressure Drop (Pa)',fontweight='bold')
plt.title('Our simulation to PNM model b/w Pressure drop and Velocity')
plt.legend(loc='best', shadow=True, fontsize='large')
plt.grid(True)


ax.annotate('most efficient catalyst', xy=(.5, .47), xytext=(.57, .47),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('least efficient catalyst', xy=(.4, 2000), xytext=(.16, 2050),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()
```
