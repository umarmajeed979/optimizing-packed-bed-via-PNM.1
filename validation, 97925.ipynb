# validation

as till now we have calculated all targeted transport properties so now our task is to validate these properties , in that regard we got transport properties data of catalysts from slumberger isl. and validate it 

##  importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

##  1st validation , porosity

```python


# Experimental and Simulation Data
catalysts_packings = ['Catalyst 1', 'Catalyst 2', 'Catalyst 3', 'Catalyst 4', 'Catalyst 5']
porosity_sim = [0.37586413, 0.30673585, 0.24561348, 0.19751556, 0.14655910]
porosity_exp = [0.44, 0.38, 0.31, 0.26, 0.21]

# Define bar width
bar_width = 0.38

# Define x-axis locations for the bars
x = np.arange(len(catalysts_packings))

# Plotting the data
plt.bar(x - bar_width/2, porosity_sim, bar_width, color='red', label='Simulated Porosity')
plt.bar(x + bar_width/2, porosity_exp, bar_width, color='blue', label='Experimental Porosity')

# Add labels and title
plt.xlabel('Catalysts Packings')
plt.ylabel('Porosity')
plt.title('Experimental Porosity Vs PNM Simulation Porosity Bars')
plt.xticks(x, catalysts_packings)  # Set x-axis ticks to be the catalysts packings
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```

## 2nd validation effective_diffusivity

```python
import matplotlib.pyplot as plt
import numpy as np

# Experimental and Simulation Data
catalysts_packings = ['Catalyst 1', 'Catalyst 2', 'Catalyst 3', 'Catalyst 4', 'Catalyst 5']
effective_diffusivity_sim = [0.25674162,0.17109867,0.10349082, 0.05633854,0.02501433]
effective_diffusivity_exp = [0.310,0.228,0.162, 0.104,0.064]

# Define bar width
bar_width = 0.38

# Define x-axis locations for the bars
x = np.arange(len(catalysts_packings))

# Plotting the data
plt.bar(x - bar_width/2, effective_diffusivity_sim, bar_width, color='red', label='Simulated Effective_Diffusivity')
plt.bar(x + bar_width/2, effective_diffusivity_exp, bar_width, color='blue', label='Experimental Effective_Diffusivity')

# Add labels and title
plt.xlabel('Catalysts Packings')
plt.ylabel('Effective_Diffusivity')
# plt.title('Experimental Effective_Diffusivity Vs PNM Simulation Effective_Diffusivity Bars')
plt.xticks(x, catalysts_packings)  
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```

## 3rd validation Pressure Drop

```python
import numpy as np
# 150kpa
u = np.arange(0.02,0.6,0.02)  
# viscosity = 0.0000179 # kg/m.s
viscosity = 1.79*10**-5 # kg/m.s
epsilon1 = 0.36
# epsilon1 = 0.428
dp = 0.00724 # m
# =1.257 
density = 1.025 # kg/m3

delta_p1 = 150 * u * viscosity * (1 - epsilon1)**2 / ((dp**2)* (epsilon1**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon1)) / (dp * epsilon1 **3)
dp1 = (delta_p1 + delta_p2) * 130/1000
print(dp1)


# catalyst 1

epsilon2 = 0.37586413 

delta_p1 = 150 * u * viscosity * (1 - epsilon2)**2 / ((dp**2)* (epsilon2**3))
delta_p2 = (1.75 * density * u**2 * (1 - epsilon2)) / (dp * epsilon2 **3)
dp2 = (delta_p1 + delta_p2) * 130/1000
print(dp2)

import matplotlib.pyplot as plt

plt.scatter(u,dp1 , color='red', marker='o' , label='Experimental Pressure Drop')
plt.scatter(u,dp2 , color='blue', marker='*' , label='Simulated Pressure Drop using PNM')
plt.xlabel('Velocity (m/s)',fontweight='bold')
plt.ylabel('Pressure Drop (Pa)',fontweight='bold')
# plt.title('Graph b/w P/L and Velocity( at 0kpa and catalyst 1)')
plt.legend(loc='best', shadow=True, fontsize='large')
plt.grid(True)
# plt.savefig('mm3.png', transparent=True)
plt.tight_layout()
plt.show()
```
