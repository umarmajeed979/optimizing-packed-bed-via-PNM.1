```python
# -- coding: utf-8 --
"""
Created on Tue Feb 27 09:57:54 2024

@author: DELL
"""

import matplotlib.pyplot as plt
import numpy as np
import openpnm as op
import pickle
import os
from matplotlib.ticker import MultipleLocator

nets = ['C:\\Users\\DELL\\Documents\\Umar_8725\\catalyst_type1_network.net','C:\\Users\\DELL\\Documents\\Umar_8725\\catalyst_type2_network.net','C:\\Users\\DELL\\Documents\\Umar_8725\\catalyst_type3_network.net','C:\\Users\\DELL\\Documents\\Umar_8725\\catalyst_type4_network.net','C:\\Users\\DELL\\Documents\\Umar_8725\\catalyst_type5_network.net']

# Define the colors and labels for each network
colors = ['brown', 'indigo', 'yellow', 'pink', 'turquoise']
# colors = ['brown', 'blue', 'green', 'red', 'purple']
Catalyst_label = ['Catalyst Type I', 'Catalyst Type II', 'Catalyst Type III', 'Catalyst Type IV', 'Catalyst Type V']

# Create the figure and axes
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Loop over each network file
for i, net_filename in enumerate(nets):
    # Construct the full file path
    file_path = os.path.join("C:\\Users\\DELL\\Documents\\Umar_8725", net_filename)

    # Load the network from file
    with open(file_path, "rb") as f:
        net = pickle.load(f)
#for i, net in enumerate(nets):
    # Load the network from file
    ##with open("D:\\OneDrive - Higher Education Commission\\After PhD\\Research\\" \
              #f"Paper7_RegionGrowing\\pore_networks\\{net}", "rb") as f:
        #net = pickle.load(f)

    # Convert to OpenPNM network
    pn = op.io.network_from_porespy(net)

    # Calculate the labels and normalized frequency of each bar for coordination number distribution
    coordination_number = pn.num_neighbors(pn.pores(), flatten=False)
    labels, counts = np.unique(coordination_number, return_counts=True)
    norm_counts = counts / np.sum(counts)

    # Plot the histogram with one bar against one minor tick in the x-axis for coordination number distribution
    ax.bar(labels, norm_counts, align='center', edgecolor=(0, 0, 0, 1), alpha=0.8, color=colors[i], label=Catalyst_label[i], width=1)
    ax.set_xlabel('b) Coordination Number', fontsize=12, fontname='Times New Roman', fontweight='bold')
    ax.set_ylabel('Normalized Frequency', fontsize=12, fontname='Times New Roman', fontweight='bold')
    ax.tick_params(axis='both', which='major', labelsize=10)
#     ax.set_title('Coordination Number Distribution', fontsize=14, fontname='Times New Roman', fontweight='bold')
#     ax.legend()
    ax.legend(loc='best', shadow=True, fontsize='large')
    ax.set_xlim(0, 35)
    ax.xaxis.set_minor_locator(MultipleLocator(1))
    ax.tick_params(axis='both', which='minor', length=3, width=1, direction='in')
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')
    ax.xaxis.set_tick_params(which='minor', top=False, direction='out')
    ax.xaxis.set_tick_params(which='major', top=True, direction='in')
    ax.yaxis.set_tick_params(which='both', right=True, direction='in')

# Set the font for all text elements in the plot
plt.rcParams['font.family'] = 'Times New Roman'

# Adjust the layout and display the plot
# plt.savefig('cnd.svg', format='svg')
# plt.savefig('cn3.jpg')
plt.tight_layout()
plt.show()
```
